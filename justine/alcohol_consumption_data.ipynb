```python
import pandas as pd
```

```python
csv_file = "resources/alcohol_consumption.csv"
alcohol_df = pd.read_csv(csv_file)

alcohol_df.head()
```

# Transform DataFrame (Clean)

```python
alcohol_new = alcohol_df.drop(columns=["Region", "Hemisphere", "HappinessScore", "HDI", "GDP_PerCapita"])



alcohol_new.head()
```

```python
countries = alcohol_new["Country"]


for country in countries:
    avg_df = alcohol_new[["Beer_PerCapita", "Spirit_PerCapita", "Wine_PerCapita"]].sum(axis=1)
    alcohol_new["Alcohol_Per_Capita"] = avg_df


alcohol_new.head()
```

```python
alcohol_new2 = alcohol_new.drop(columns=["Beer_PerCapita", "Spirit_PerCapita", "Wine_PerCapita"])

alcohol_new2.head()
```

```python
alcohol_new2.set_index("Country", inplace=True)

alcohol_new2.head()
```

```python
# Save clean data to output file

alcohol_new2.to_csv("alcohol_csv", index=True)
```

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "resources/world_happiness.csv"
happiness_df = pd.read_csv(csv_file)
happiness_df.head()
```

```python
# Drop columns that are not needed

happiness_df_new = happiness_df.drop(columns=["Social support", "Freedom to make life choices", "Generosity", "Perceptions of corruption"])

happiness_df_new

```

```python
# Rename columns & reset index

clean_happiness_df = happiness_df_new.rename(columns={"Overall rank": "Overall Happiness Rank",
                                                     "Score":"Overall Happiness Score",
                                                      "Country or region": "Country or Region",
                                                     "GDP per capita":"GDP Per Capita",
                                                     "Healthy life expectancy": "Life Expectancy"})

clean_happiness_df.head()
```

```python
clean_happiness_df.set_index("Country or Region", inplace=True)
```

```python
clean_happiness_df
```

```python
# Save clean data to output file

clean_happiness_df.to_csv("happiness_csv", index=False)
```
